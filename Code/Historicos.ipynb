```python
import pandas as pd
import os

# Ruta de la carpeta que contiene los archivos CSV
carpeta = os.path.join(os.getcwd(), 'unificados')

# Lista para almacenar los DataFrames de cada archivo
dataframes = []

# Iterar sobre los archivos en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith('.csv'):
        ruta_completa = os.path.join(carpeta, archivo)
        
        # Leer el archivo CSV
        df = pd.read_csv(ruta_completa, encoding='ISO-8859-1')
        
        # Dividir la columna 'Datos'
        df[['Fecha', 'Precio']] = df['Datos'].str.split(', ', expand=True)
        
        # Eliminar la columna 'Datos'
        df = df.drop('Datos', axis=1)
        
        # Agregar corchetes a las fechas y precios
        df['Fecha'] = df['Fecha'].apply(lambda x: f"['{x}']")
        df['Precio'] = df['Precio'].apply(lambda x: f"['{x}']")

        df['Fecha'] = df['Fecha'].str.replace(r'[\[\]\'\']', '', regex=True)
        df['Precio'] = df['Precio'].str.replace(r'[\[\]\'\']', '', regex=True)
        
        # Agregar el DataFrame a la lista
        dataframes.append(df)

# Concatenar todos los DataFrames en uno
resultado_final = pd.concat(dataframes, ignore_index=True)

# Guardar el resultado final en un único archivo CSV
resultado_final.to_csv('extracciones_columna1.csv', index=False)
```
